# Data Exploration


# Question 1: Which state has the most counties in it? 

```python
df.head()
```

```python
# Count the number of counties in each state
county_counts = df.groupby('STNAME').size()
print(county_counts)
```

```python
#Find the maximum county
max_counties = county_counts.max()
print(max_counties)
```

```python
# Find the state with the maximum number of counties
state_with_max_counties = county_counts[county_counts == max_counties].index[0]
print(state_with_max_counties)
```

```python
print(f'The state with the most counties is {state_with_max_counties} with {max_counties} counties.')
```

# Question 2: Only looking at the three most populous counties for each state, what are the three most populous states (in order of highest population to lowest population)? Using CENSUS2010POP

```python
# Convert the population column to integers
df['CENSUS2010POP'] = df['CENSUS2010POP'].astype(int)
df['CENSUS2010POP'].describe()
```

```python
# Group the data by state and sum the population for each state
populations = df.groupby('STNAME')['CENSUS2010POP'].sum()
populations
```

```python
# Sort the data by population in descending order
populations = populations.sort_values(ascending=False)
populations
```

```python
# Get the top 3 most populous states
top_3_states = populations.sort_values(ascending=False).head(3)
top_3_states
```

```python
# Print the result in order of highest to lowest population
print(f'The three most populous states are: {top_3_states.index[0]} ({top_3_states[0]:,}), {top_3_states.index[1]} ({top_3_states[1]:,}), and {top_3_states.index[2]} ({top_3_states[2]:,}).')
```

# Question 3: Which City has the most counties in it?

```python
# Count the number of counties in each city
county_counts_city = df.groupby('CTYNAME').size()
print(county_counts_city)
```

```python
#Find the maximum county number in a city
max_counties_city = county_counts_city.max()
print(max_counties_city)
```

```python
# Find the city with the maximum number of counties
city_with_max_counties = county_counts_city[county_counts_city == max_counties_city].index[0]
print(city_with_max_counties)
```

```python
print(f'The city with the most counties is {city_with_max_counties} with {max_counties_city} counties.')
```

# Question 4: Which region has the most division in it?

```python
# Count the number of counties in each state
division_counts = df.groupby('REGION').size()
print(division_counts)
```

```python
#Find the maximum division
max_division = division_counts.max()
print(max_division)
```

```python
# Find the Region with the maximum number of divisions
region_with_max_division = division_counts[division_counts == max_division].index[0]
print(region_with_max_division)
```

```python
print(f'The region with the most divisions is Region {region_with_max_division} with {max_division} divsions.')
```
